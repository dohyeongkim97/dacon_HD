```python
def colomn_processor(df):
    if 'SAMPLE_ID' in df.columns:
        df = df.drop(columns = ['SAMPLE_ID'])
    else:
        pass
    
    if 'ID' in df.columns:
        df = df.drop(columns = ['ID'])
    
    if 'SHIPMANAGER' in df.columns:
        df = df.drop(columns = ['SHIPMANAGER'])
    
    if isinstance(df.loc[len(df)-1, 'ATA'], str):
        df['ATA'] = pd.to_datetime(df['ATA'])
            
            
    return df
```

```python
df = colomn_processor(df)
test = colomn_processor(test)
```

```python
ohe = OneHotEncoder(drop='first', handle_unknown='ignore', min_frequency=3, sparse=False,)

def encoder(df, train=True):
    if train:
        ohe.fit(df[['ARI_CO', 'ARI_PO', 'SHIP_TYPE_CATEGORY', 'FLAG']])
        global encoded
        encoded = pd.DataFrame(ohe.transform(df[['ARI_CO', 'ARI_PO', 'SHIP_TYPE_CATEGORY', 'FLAG']]), columns=ohe.get_feature_names_out())
        df.drop(columns = ['ARI_CO', 'ARI_PO', 'SHIP_TYPE_CATEGORY', 'FLAG'], inplace=True)
        
        df['month'] = df['ATA'].dt.month
        df['year'] = df['ATA'].dt.year
        df['calc_month'] = (df['year']-2014)*12 + df['month']
        
        df.drop(columns = ['ATA'], inplace=True)
        
        return df
        
    else:
        global encoded_test
        encoded_test = pd.DataFrame(ohe.transform(df[['ARI_CO', 'ARI_PO', 'SHIP_TYPE_CATEGORY', 'FLAG']]), columns=ohe.get_feature_names_out())
        df.drop(columns = ['ARI_CO', 'ARI_PO', 'SHIP_TYPE_CATEGORY', 'FLAG'], inplace=True)
        
        df['month'] = df['ATA'].dt.month
        df['year'] = df['ATA'].dt.year
        df['calc_month'] = (df['year']-2014)*12 + df['month']
        
        df.drop(columns = ['ATA'], inplace=True)
        
        return test
```
